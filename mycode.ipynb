```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras import regularizers
```

```python
df=pd.read_csv('/content/housing.csv')
df.drop(columns=['ocean_proximity'],inplace=True)
df.head()
```

```python
x=df[df.columns[:-1]].values
y=df[df.columns[-1]].values
```

```python
scaler_x = StandardScaler()
scaler_y = StandardScaler()
x_scaled = scaler_x.fit_transform(x)
y_scaled = scaler_y.fit_transform(y.reshape(-1, 1))
```

```python
xtrain,xtemp,ytrain,ytemp=train_test_split(x_scaled,y_scaled,test_size=0.4,random_state=0)
xval,xtest,yval,ytest=train_test_split(xtemp,ytemp,test_size=0.5,random_state=0)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001)),
    tf.keras.layers.Dense(32, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001)),
    tf.keras.layers.Dense(1)
])
```

```python
model.compile(optimizer='adam',loss='mean_squared_error',metrics=[])
```

```python
history = model.fit(xtrain, ytrain, epochs=100, batch_size=32, validation_data=(xval, yval))
```

```python
test_loss = model.evaluate(xtest, ytest)
print(f'Test Loss: {test_loss:.4f}')
```
